```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = glob.glob('../train_images/*') # write the address here of the images

print(len(images))

print(images[0:10])
print(type(images[0]))
```

```python
print(images[0][29:])
```

```python
images_train = []
for x in range(200):
    images_train.append(cv2.imread(images[x]))
```

```python
images_train = np.array(images_train, dtype=object)
print(np.shape(images_train))
```

```python
#scale_percent = 60 # percent of original size
#width = int(img.shape[1] * scale_percent / 100)
#height = int(img.shape[0] * scale_percent / 100)

width = 256
height = 256

dim = (width, height)
  
# resize image
resized = []
for x in range(100):
    resized.append(cv2.resize(images_train[x], dim, interpolation = cv2.INTER_AREA))

#cv2.imshow("Resized image", resized[0])
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10),
                                    sharex=True, sharey=True)
ax1.imshow(images_train[1], cmap='gray')
ax1.axis('off')
ax1.set_title('Original')
ax2.imshow(resized[1], cmap='magma')
ax2.axis('off')
ax2.set_title('Resized')

fig.tight_layout()
plt.show()
```
